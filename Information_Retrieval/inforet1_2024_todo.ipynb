# Tokenization

### 🚧 todo:
Describe in one sentence the difference between the tokenizations. Which one is your favorite and why?


```python
# 1.
text = "Let's all together defeat last year's problem, SARS-CoV-2, in 2022!"
'|'.join(text.split()),len(text.split())
```

```python
# 2.
nochar = re.compile('\W+')
'|'.join(nochar.split(text)),len(nochar.split(text))
```

```python
# 3.
nochar = re.compile('(\W+)')
'|'.join(nochar.split(text)),len(nochar.split(text))
```

**Analysis:**
- The first tokenization seperates texts by whitespace.
- The second tokenization splits texts by non-alphabet and non-numerical characters  and only preserve alphabets and numbers.
- The third tokenization follows the same tokenization rules but preserve other characters as tokens.

I prefer the fourth one because it allows for the preservation of relevant information, such as hyphenated words or alphanumeric identifiers, while still separating the text into individual tokens.

```python
#4.
charorhyphen = re.compile(r'[\w-]+')
'|'.join(charorhyphen.findall(text)),len(charorhyphen.findall(text))
```

### Let's first flatten the list

```python
alltoks = [item for sublist in df['tokens'] for item in sublist]
len(alltoks)
```

```python
text = "Let's all together defeat last year's problem, SARS-CoV-2, in 2021!"
tokens = word_tokenize(text)
counter = Counter(tokens)
counter
```

### What are the most common words of English?

```python
counter = Counter(alltoks)
counter.most_common(22)
```

for even bigger databases, it might be advisable to do the computation iteratively:

```python
counter = Counter()
df['tokens'].map(counter.update)
counter.most_common(22)
```

```python
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df.sort_values('freq',  inplace=True, ascending=False)
freq_df
```

# a big vocabulary:
grab a pageview file here https://dumps.wikimedia.org/other/pageviews/2022/2022-01/

we produce a list of potential terms from it:

```python
terms = []
for li in open('/content/gdrive/MyDrive/Colab Notebooks/Information Retrieval/pageviews-20220101-000000').read().strip().split('\n'):
    t=li.split()[1]
    if li[:2]=='en' and t[:5]!='File:':
        if t[:9]=='Category:':
            t=t[9:] # can be improved Page:, Template:, ...
        terms+=[t.replace('_',' ')]
terms = sorted(set(terms))
open('en.pages.txt','w').write('\n'.join(terms))
```

```python
Counter([len(t.split()) for t in terms]).most_common()
```

```python
[t for t in terms if len(t.split())>33]
```

## find the most frequently encountered Wikipedia entity

In number of speeches:


    正确答案是the

```python
# change it in matrix form
X_mod = X.toarray()
X_mod[X_mod > 0] = 1

freq_speech = np.sum(X_mod, axis=0)
```

```python
index = freq_speech.argmax()
print(f"The most frequent entity in number of speeches is '{terms[index]}' with '{max(freq_speech)}' appearances")
```

In number of occurrences

```python
X_mod = X.toarray()

freq_occurances = np.sum(X_mod, axis=0)
```

```python
index = freq_occurances.argmax()
print(f"The most frequent entity in number of occurances is '{terms[index]}' with '{max(freq_occurances)}' appearances")
```

Which speech talks most about the 'Union of African States'?

```python
X_list = X_mod[:, terms.index('Union of African States')]
id_tar = X_list.argmax()

print(f"The speech ID {id_tar} talks most about the 'Union of African States")
```
